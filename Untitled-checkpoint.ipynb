```python
x_train,x_test,y_train,y_test=train_test_split(x,y_normalized,test_size=0.2,random_state=40)
```

```python
from sklearn.decomposition import PCA
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)       # Transform test data using the same scaler

# Apply PCA
pca = PCA(n_components=2)
X_train = pca.fit_transform(x_train)  # Fit and transform on training data
X_test = pca.transform(X_test) 
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Define empty lists to store metrics
mse_list = []
mae_list = []
r2_list = []
# Function to calculate various metrics such as accuracy, precision etc
def calculatemetrics(algorithm, predict, testY):
    testY = testY.astype('int')
    predict = predict.astype('int')
    
    mse = mean_squared_error(testY, predict)
    mae = mean_absolute_error(testY, predict)
    r2 = r2_score(testY, predict) * 100
    
    # Append calculated metrics to lists
    mse_list.append(mse)
    mae_list.append(mae)
    r2_list.append(r2)
    
    print(algorithm + ' Mean Squared Error: ' + str(mse))
    print(algorithm + ' Mean Absolute Error: ' + str(mae))
    print(algorithm + ' R^2 Score: ' + str(r2))
    
    # Ensure that you are working with numpy arrays
    testY = np.array(testY)
    predict = np.array(predict)

    # Create a scatter plot
    plt.figure(figsize=(8, 8))
    plt.scatter(predict, testY,color='blue')
    plt.plot([np.min(testY), np.max(testY)], [np.min(testY), np.max(testY)], linestyle='--', color='red', lw=2)
    # Set labels and title
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.title('Regression Performance')
    plt.show()
```

```python
model_folder = 'model'
path = os.path.join(model_folder, 'ridge_model.pkl')
if os.path.exists(path):
    # Load the model if it exists
    print("Loading existing model...")
    mod1 = joblib.load(path)
    predict=mod1.predict(x_test)
    calculatemetrics("RidgeRegression",predict,y_test)
else:
    # Train a new model if it doesn't exist
    print("Training a new Decision Tree model...")
    mod1 = Ridge()
    mod1.fit(x_train, y_train)
    # Save the model to a file for future use
    joblib.dump(mod1, path)
    print(f"Model saved to {path}")
    y_pred = mod1.predict(x_test)
    calculatemetrics("RidgeRegression",y_pred,y_test)
```

```python
model_folder = 'model'
path = os.path.join(model_folder, 'lg.pkl')
if os.path.exists(path):
    # Load the model if it exists
    print("Loading existing model...")
    model = joblib.load(path)
    predict=model.predict(x_test)
    calculatemetrics("linearRegression",predict,y_test)
else:
    # Train a new model if it doesn't exist
    print("Training a new Decision Tree model...")
    model = Lasso()
    model.fit(x_train, y_train)
    # Save the model to a file for future use
    joblib.dump(model, path)
    print(f"Model saved to {path}")
    y_pred = model.predict(x_test)
    calculatemetrics("linearRegression",y_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(25,)))
model.add(Dense(32, activation='relu'))  # You will extract features from this layer
model.add(Dense(1))  # Final output layer (for classification)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the DNN
model.fit(x_train, y_train, epochs=10, verbose=0)

# Step 2: Feature extraction using the trained DNN (extracting the output from the second-to-last layer)
extractor = Sequential(model.layers[:-1])  # Remove the output layer for feature extraction
X_train_features = extractor.predict(x_train)
X_test_features = extractor.predict(x_test)

rfc =RandomForestRegressor()
rfc.fit(X_train_features, y_train)


# Step 4: Make predictions on test data and evaluate
y_pred = rfc.predict(X_test_features)

# Accuracy
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'MSE: {mse}, MAE: {mae}, R²: {r2}')
```
